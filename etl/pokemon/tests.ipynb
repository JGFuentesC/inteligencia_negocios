```python
import requests
import json
import os
from glom import glom 
import pandas as pd
pd.set_option('display.max_columns', None)  
pd.set_option('display.max_colwidth', False)
```

```python
def consultarPokemon(num:int):
    url = f"https://pokeapi.co/api/v2/pokemon/{num}"
    response = requests.get(url)
    data = response.json()

    with open(os.path.join('./data',f'{num:03d}.json'), "w") as file:
        json.dump(data, file)
        file.close()
```

```python
def parsePokemon(num:int)->pd.DataFrame:
    claves = ['height',
    'id',
    'sprites.other.official-artwork.front_default',
    'stats',
    'weight',
    'name']
    data = json.load(open(f'./data/{num:03d}.json'))
    data = pd.Series(dict(zip(claves, map(lambda k:glom(data,k),claves))))
    data = data.to_frame().T
    def splitStats(stats:list)->tuple:
        return {glom(st,'stat.name'):glom(st,'base_stat') for st in stats}
    data['stats'] = data['stats'].map(splitStats) 
    return pd.concat([data.drop('stats',axis=1),data['stats'].apply(pd.Series)],axis=1)
```

```python
df = pd.concat(map(parsePokemon,range(1,152)),ignore_index=True)
```
